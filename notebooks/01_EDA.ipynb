TARGET VARIABLE ANALYSIS

```python
df["Class"].value_counts()
```

```python
fraud_count = df["Class"].sum()
total_count = len(df)

fraud_percentage = fraud_count/total_count * 100

print(f"Fraud Trasnsactions: {fraud_count}")
print(f"Total Transactions: {total_count}")
print(f"Fraud Percentage: {fraud_percentage:.4f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))

sns.countplot(x="Class", data=df)

plt.title("Fraud vs Non-Fraud Transactions")

plt.show()
```

The dataset is highly imbalanced.

Only about 0.17% of transactions are fraudulent,
making fraud detection a challenging classification problem.

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Amount"], bins=50) #splits amount into 50 ranges and show how many fall into each range.
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.show()
```

```python
df["Amount"].describe()
```

FRAUD vs AMT

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x="Class",
    y="Amount",
    data = df
)
plt.title("Transaction Amount by Class")
plt.show()
```

- Most transactions in both classes are of low value.
- Non-fraudulent transactions have a wider range of amounts(smaller to very large).
- Fraudulent transactions are generally of smaller amounts.
- The `Amount` feature is highly right-skewed due to the presence of large outliers.

TIME ANALYSIS

```python
df["Hour"] = (df["Time"] // 3600) % 24  # Extract hour from Time column

plt.figure(figsize=(12,5))
sns.countplot(x=df["Hour"])
plt.title("Transactions by Hour")
plt.show()
```

```python
fraud_by_hour = df.groupby("Hour")["Class"].mean() # Calculate fraud rate by hour

fraud_by_hour.plot(
    kind = "bar",
    figsize = (12,5),
)
plt.title("Fraud Rate by Hour")
plt.ylabel("Fraud Rate")
plt.show()

#percentage (rate) of transactions that are fraudulent in each hour.
```

CORRELATION ANALYSIS

```python
corr_with_target = (
    df.corr(numeric_only = True)["Class"]
    .sort_values()
)

corr_with_target
```

```python
print("Most Negative Correlations")
print(corr_with_target.head(10))

print("\nMost Positive Correlations")
print(corr_with_target.tail(10))
```

V11 = 0.155 means higher values of V11 are slightly associated with fraud.

V17 = -0.326 means higher values of V17 are slightly associated with non-fraud.

CORRELATION HEATMAP

```python
plt.figure(figsize = (14,10))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap = "coolwarm",
)

plt.title("Feature Correlation Heatmap")
plt.show()
```
